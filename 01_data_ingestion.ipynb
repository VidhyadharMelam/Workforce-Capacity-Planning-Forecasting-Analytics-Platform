# Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
```

**Loading the Dataset with Pandas**

```python
# Step 3: Define file paths
oews_path = "oews.csv"
employment_path = "employment_industry.csv"
labor_force_path = "labor_force.csv"
```

# Step 2: Load Datasets

```python
# Step 4: Load datasets into Pandas
oews = pd.read_csv(oews_path)
employment = pd.read_csv(employment_path)
labor_force = pd.read_csv(labor_force_path)
```

Step 1: Inspect the Dataset Structure

```python
# Step 5: Confirm dataset shapes
print("OEWS Shape:", oews.shape)
print("Employment Shape:", employment.shape)
print("Labor Force Shape:", labor_force.shape)
```

```python
# Step 6: Preview first few rows
display(oews.head())
display(employment.head())
display(labor_force.head())
```

Metadata Summary

```python
# Step 7: Metadata summary
print("\nOEWS Info:")
print(oews.info())

print("\nEmployment Info:")
print(employment.info())

print("\nLabor Force Info:")
print(labor_force.info())
```

Save Raw Copies (Optional)

```python
# Step 8: Save raw copies (optional)

import os

# Create the results folder if it doesn't exist
os.makedirs("results", exist_ok=True)

# Save raw copies of each dataset
oews.to_csv("results/oews_raw_copy.csv", index=False)
employment.to_csv("results/employment_raw_copy.csv", index=False)
labor_force.to_csv("results/labor_force_raw_copy.csv", index=False)
```
